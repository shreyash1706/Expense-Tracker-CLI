```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
expense_f="example2.csv"
```

```python
df=pd.read_csv(expense_f)

df['Date']=pd.to_datetime(df.Date,format='%d-%m-%Y')
df['month']=df['Date'].dt.month
df['year']=df['Date'].dt.year
TotalExpenseByMonth=df.groupby(['year','month']).Amount.sum()


avgPerDay=df.groupby(['year','month']).Amount.mean()
grouped_series=df.groupby(['year','month','Category']).Amount.sum()/TotalExpenseByMonth
df = grouped_series.reset_index()

pivot_df = df.pivot_table(index=['year', 'month'], columns='Category', values='Amount', fill_value=0)


pivot_df.reset_index(inplace=True)


pivot_df.columns.name = None

total_expenses_df = TotalExpenseByMonth.reset_index()
total_expenses_df.columns = ['year', 'month', 'TotalExpense']
avg_df=avgPerDay.reset_index()
avg_df.columns=['year', 'month', 'AvgPerDay']

# Merge the total expenses DataFrame with the existing pivot table DataFrame
pivot_df = pd.merge(pivot_df, total_expenses_df, on=['year', 'month'], how='left')
pivot_df = pd.merge(pivot_df, avg_df, on=['year', 'month'], how='left')


pivot_df['perv_month_expense']=pd.concat([pivot_df.TotalExpense,pd.Series([0])],axis=0).shift(1)[:-1].fillna(0)

pivot_df['next_month_expense']=pd.concat([pivot_df.TotalExpense,pd.Series([0])],axis=0).shift(-1)[:-1]
pivot_df.drop(pivot_df.head(1).index,inplace=True) 
pivot_df.drop(pivot_df.tail(1).index,inplace=True)
pivot_df
```

```python
model=RandomForestRegressor(random_state=0)
```

```python
y=pivot_df['next_month_expense']
X=pivot_df[['month', 'Food', 'Fun', 'HealthCare', 'Home', 'Others',
       'PersonalCare', 'Transport', 'Work', 'TotalExpense', 'AvgPerDay',
       'perv_month_expense']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.2)
```

```python
model.fit(X_train,y_train)
```

```python
preds=model.predict(X_test)
```

```python
mse=mean_squared_error(y_test,preds)
print(f"Mean Squared Error: {mse}")
```
